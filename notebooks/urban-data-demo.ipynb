# Urban Data Analysis - Indicadores Sinteticos

Exemplo de leitura inicial para priorizacao territorial.

```python
import pandas as pd
df = pd.read_csv('../data/indicadores_urbanos_exemplo.csv')
df['score_atencao'] = df['indice_vulnerabilidade'] * 100 + (60 - df['acesso_transporte_pct']) * 0.4 + df['tempo_medio_deslocamento_min'] * 0.3
df.sort_values('score_atencao', ascending=False)[['bairro','prioridade_intervencao','score_atencao']]
```
